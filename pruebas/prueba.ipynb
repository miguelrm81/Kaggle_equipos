```python
import pandas as pd
import numpy as np
import cv2
```

```python
df = pd.read_csv('train_set.csv')
df
```

```python
test = pd.read_csv('test_set.csv')
test
```

```python
df['label'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df['label_encoded'] = label_encoder.fit_transform(df['label'])
```

```python
df['path'] = df['path'].str.replace('\\', '/')
df['path'] = df['path'].str.replace('../data/images/train/', './data/images/train/')

def cargar_y_convertir_imagen(ruta_imagen):
    image = cv2.imread(ruta_imagen, 0)
    image_array = np.array(image)
    return image_array

df['imagenes'] =df['path'].apply(cargar_y_convertir_imagen)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
def estandarizar_lista(lista):
    array = np.array(lista)
    scaler = StandardScaler()
    array_estandarizado = scaler.fit_transform(array)
    return array_estandarizado.tolist()
df['imagenes'] = df['imagenes'].apply(estandarizar_lista)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df= df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df
```

```python
df['imagenes'] = df['imagenes'].apply(lambda x: np.array(x).flatten())

X = np.vstack(df['imagenes'].to_numpy())
y = df['label_encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

modelo = SVC(gamma='auto', cache_size=3000)
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

precision = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {precision}')
```

```python
test['path'] = test['id_img'].apply(lambda x: f'./data/images/test/{x}.jpg')
```

```python
test
```

```python
test['imagenes'] = test['path'].apply(cargar_y_convertir_imagen)
```

```python
test['imagenes'] = test['imagenes'].apply(lambda x: np.array(x).flatten())

X_test_set = np.vstack(test['imagenes'].to_numpy())

y_pred_test = modelo.predict(X_test_set)
```

```python
submission = pd.DataFrame({'id_img': test['id_img'], 'label': y_pred_test})

submission.to_csv('submission.csv', index=False)
```
